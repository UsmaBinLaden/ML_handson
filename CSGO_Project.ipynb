```python
#Label encoding 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["map"] = le.fit_transform(df["map"])
df["bomb_planted"] = le.fit_transform(df["bomb_planted"])
df["round_winner"] = le.fit_transform(df["round_winner"])
```

```python
df.info()
```

```python
# data spliting in ternms of dependent and independednt
```

```python
x = df.iloc[:,:-1]
y = df["round_winner"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 34)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train,y_train)
```

```python
lda.transform(x_test)
```

```python
lda.coef_
```

```python
lda_coef = np.exp(np.abs(lda.coef_))
lda_coef
```

```python
feature_name = x.columns
feature_name
```

```python
df_features_imp = pd.DataFrame({"Feature":feature_name ,"imp_values":lda_coef[0]})   # flatten() vi use kar sakte
```

```python
df_features_imp
```

```python
lda_coef.shape
```

```python
feature_name.shape
```

```python
top_20_fea = df_features_imp.nlargest(20,"imp_values")
#nlargest(n,colname) ---> used to get top n values on colname
top_20_fea
```

```python
imp_col = top_20_fea.index
imp_col
```

```python
x_train[:,[8,17,7]]
```

```python
x_train = x_train[:,imp_col]
x_train
```

```python
x_test = x_test[:,imp_col]
x_test
```

# logistic regression -75

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lg = LogisticRegression()
model_lg.fit(x_train,y_train)
```

```python
y_pred = model_lg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```
